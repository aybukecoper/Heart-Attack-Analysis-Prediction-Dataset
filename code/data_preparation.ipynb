```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = r"C:\Users\Lenovo\Desktop\Heart-Attack-Analysis-Prediction-Dataset\data\heart.csv"
df = pd.read_csv(path)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python

plt.figure(figsize=(20, 15))   

for i, column in enumerate(df.columns):
    plt.subplot(5, 4, i + 1)   
    df[df["output"] == 0][column].hist(bins=35, color='blue', label='Heart Disease = NO', alpha=0.6)
    df[df["output"] == 1][column].hist(bins=35, color='red', label='Heart Disease = YES', alpha=0.6)
    plt.legend(loc='upper right', frameon=True)
    plt.xlabel(column)
    
    
```

```python
# Exclude non-numeric columns
numeric_columns = df.select_dtypes(include=[np.number]).columns

# Plot linear regression line for each numeric column
for column in numeric_columns:
    sns.lmplot(x=column, y='output', data=df)
```
